```python
import numpy as np

def calculate_averages(filename):
    # 데이터 저장용 딕셔너리
    data = {
        'bonafide': {'probs': [], 'uncertainties': []},
        'spoof_7': {'probs': [], 'uncertainties': []},
        'spoof_8': {'probs': [], 'uncertainties': []},
        'spoof_9': {'probs': [], 'uncertainties': []},
        'spoof_10': {'probs': [], 'uncertainties': []},
        'spoof_11': {'probs': [], 'uncertainties': []},
        'spoof_12': {'probs': [], 'uncertainties': []},
        'spoof_13': {'probs': [], 'uncertainties': []},
        'spoof_14': {'probs': [], 'uncertainties': []},
        'spoof_15': {'probs': [], 'uncertainties': []},
        'spoof_16': {'probs': [], 'uncertainties': []},
        'spoof_17': {'probs': [], 'uncertainties': []},
        'spoof_18': {'probs': [], 'uncertainties': []},
        'spoof_19': {'probs': [], 'uncertainties': []},

    }

    # 파일 읽기
    with open(filename, 'r') as file:
        for line in file:
            parts = line.split()
            if len(parts) < 5:
                continue  # 잘못된 형식의 줄 건너뛰기

            category = parts[2]
            prob = float(parts[3])
            uncertainty = float(parts[4])
            spoof_t = parts[1]
            
            if category == 'bonafide':
                data['bonafide']['probs'].append(prob)
                data['bonafide']['uncertainties'].append(uncertainty)
            elif category == 'spoof':
                if spoof_t == 'A07':
                    data['spoof_7']['probs'].append(prob)
                    data['spoof_7']['uncertainties'].append(uncertainty)
                if spoof_t == 'A08':
                    data['spoof_8']['probs'].append(prob)
                    data['spoof_8']['uncertainties'].append(uncertainty)
                if spoof_t == 'A09':
                    data['spoof_9']['probs'].append(prob)
                    data['spoof_9']['uncertainties'].append(uncertainty)
                if spoof_t == 'A10':
                    data['spoof_10']['probs'].append(prob)
                    data['spoof_10']['uncertainties'].append(uncertainty)
                if spoof_t == 'A11':
                    data['spoof_11']['probs'].append(prob)
                    data['spoof_11']['uncertainties'].append(uncertainty)
                if spoof_t == 'A12':
                    data['spoof_12']['probs'].append(prob)
                    data['spoof_12']['uncertainties'].append(uncertainty)
                if spoof_t == 'A13':
                    data['spoof_13']['probs'].append(prob)
                    data['spoof_13']['uncertainties'].append(uncertainty)
                if spoof_t == 'A14':
                    data['spoof_14']['probs'].append(prob)
                    data['spoof_14']['uncertainties'].append(uncertainty)
                if spoof_t == 'A15':
                    data['spoof_15']['probs'].append(prob)
                    data['spoof_15']['uncertainties'].append(uncertainty)
                if spoof_t == 'A16':
                    data['spoof_16']['probs'].append(prob)
                    data['spoof_16']['uncertainties'].append(uncertainty)
                if spoof_t == 'A17':
                    data['spoof_17']['probs'].append(prob)
                    data['spoof_17']['uncertainties'].append(uncertainty)
                if spoof_t == 'A18':
                    data['spoof_18']['probs'].append(prob)
                    data['spoof_18']['uncertainties'].append(uncertainty)
                if spoof_t == 'A19':
                    data['spoof_19']['probs'].append(prob)
                    data['spoof_19']['uncertainties'].append(uncertainty)

    # 평균 계산 함수
    def calculate_mean(values):
        return np.mean(values) if values else 0

    # bonafide 평균 계산
    bonafide_prob_avg = calculate_mean(data['bonafide']['probs'])
    bonafide_uncertainty_avg = calculate_mean(data['bonafide']['uncertainties'])
    spoof_7_prob_avg = calculate_mean(data['spoof_7']['probs'])
    spoof_7_uncertainty_avg = calculate_mean(data['spoof_7']['uncertainties'])
    spoof_8_prob_avg = calculate_mean(data['spoof_8']['probs'])
    spoof_8_uncertainty_avg = calculate_mean(data['spoof_8']['uncertainties'])
    spoof_9_prob_avg = calculate_mean(data['spoof_9']['probs'])
    spoof_9_uncertainty_avg = calculate_mean(data['spoof_9']['uncertainties'])
    spoof_10_prob_avg = calculate_mean(data['spoof_10']['probs'])
    spoof_10_uncertainty_avg = calculate_mean(data['spoof_10']['uncertainties'])
    spoof_11_prob_avg = calculate_mean(data['spoof_11']['probs'])
    spoof_11_uncertainty_avg = calculate_mean(data['spoof_11']['uncertainties'])
    spoof_12_prob_avg = calculate_mean(data['spoof_12']['probs'])
    spoof_12_uncertainty_avg = calculate_mean(data['spoof_12']['uncertainties'])
    spoof_13_prob_avg = calculate_mean(data['spoof_13']['probs'])
    spoof_13_uncertainty_avg = calculate_mean(data['spoof_13']['uncertainties'])
    spoof_14_prob_avg = calculate_mean(data['spoof_14']['probs'])
    spoof_14_uncertainty_avg = calculate_mean(data['spoof_14']['uncertainties'])
    spoof_15_prob_avg = calculate_mean(data['spoof_15']['probs'])
    spoof_15_uncertainty_avg = calculate_mean(data['spoof_15']['uncertainties'])
    spoof_16_prob_avg = calculate_mean(data['spoof_16']['probs'])
    spoof_16_uncertainty_avg = calculate_mean(data['spoof_16']['uncertainties'])
    spoof_17_prob_avg = calculate_mean(data['spoof_17']['probs'])
    spoof_17_uncertainty_avg = calculate_mean(data['spoof_17']['uncertainties'])
    spoof_18_prob_avg = calculate_mean(data['spoof_18']['probs'])
    spoof_18_uncertainty_avg = calculate_mean(data['spoof_18']['uncertainties'])
    spoof_19_prob_avg = calculate_mean(data['spoof_19']['probs'])
    spoof_19_uncertainty_avg = calculate_mean(data['spoof_19']['uncertainties'])
    

    # 결과 출력
    print(f"Bonafide - Probability Average: {bonafide_prob_avg:.4f}")
    print(f"Bonafide - Uncertainty Average: {bonafide_uncertainty_avg:.4f}")
    print(f"A07 - Probability Average: {spoof_7_prob_avg:.4f}")
    print(f"A07 - Uncertainty Average: {spoof_7_uncertainty_avg:.4f}")
    print(f"A08 - Probability Average: {spoof_8_prob_avg:.4f}")
    print(f"A08 - Uncertainty Average: {spoof_8_uncertainty_avg:.4f}")
    print(f"A09 - Probability Average: {spoof_9_prob_avg:.4f}")
    print(f"A09 - Uncertainty Average: {spoof_9_uncertainty_avg:.4f}")
    print(f"A10 - Probability Average: {spoof_10_prob_avg:.4f}")
    print(f"A10 - Uncertainty Average: {spoof_10_uncertainty_avg:.4f}")
    print(f"A11 - Probability Average: {spoof_11_prob_avg:.4f}")
    print(f"A11 - Uncertainty Average: {spoof_11_uncertainty_avg:.4f}")
    print(f"A12 - Probability Average: {spoof_12_prob_avg:.4f}")
    print(f"A12 - Uncertainty Average: {spoof_12_uncertainty_avg:.4f}")
    print(f"A13 - Probability Average: {spoof_13_prob_avg:.4f}")
    print(f"A13 - Uncertainty Average: {spoof_13_uncertainty_avg:.4f}")
    print(f"A14 - Probability Average: {spoof_14_prob_avg:.4f}")
    print(f"A14 - Uncertainty Average: {spoof_14_uncertainty_avg:.4f}")
    print(f"A15 - Probability Average: {spoof_15_prob_avg:.4f}")
    print(f"A15 - Uncertainty Average: {spoof_15_uncertainty_avg:.4f}")
    print(f"A16 - Probability Average: {spoof_16_prob_avg:.4f}")
    print(f"A16 - Uncertainty Average: {spoof_16_uncertainty_avg:.4f}")
    print(f"A17 - Probability Average: {spoof_17_prob_avg:.4f}")
    print(f"A17 - Uncertainty Average: {spoof_17_uncertainty_avg:.4f}")
    print(f"A18 - Probability Average: {spoof_18_prob_avg:.4f}")
    print(f"A18 - Uncertainty Average: {spoof_18_uncertainty_avg:.4f}")
    print(f"A19 - Probability Average: {spoof_19_prob_avg:.4f}")
    print(f"A19 - Uncertainty Average: {spoof_19_uncertainty_avg:.4f}")
    
# 파일명에 맞는 경로를 지정하세요
filename = '/home/jykang/models/aasist/exp_result_evid/LA_AASIST_debug_ep100_bs24/eval_scores_using_best_dev_model.txt'
calculate_averages(filename)
```
