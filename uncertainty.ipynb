```python
import numpy as np

def calculate_averages(filename):
    spoof_keys = [f'spoof_{i}' for i in range(7, 20)]
    data = {key: {'probs': [], 'uncertainties': []} for key in ['bonafide'] + spoof_keys}

    # 파일 읽기 및 데이터 저장
    with open(filename, 'r') as file:
        for line in file:
            parts = line.split()
            if len(parts) < 5:
                continue

            category = parts[2]
            prob = float(parts[3])
            uncertainty = float(parts[4])
            spoof_t = parts[1]

            if category == 'bonafide':
                key = 'bonafide'
            elif category == 'spoof' and spoof_t.startswith('A'):
                key = f'spoof_{int(spoof_t[1:]):d}'
            else:
                continue

            data[key]['probs'].append(prob)
            data[key]['uncertainties'].append(uncertainty)

    # 평균 계산 및 결과 출력
    for key in data:
        prob_avg = np.mean(data[key]['probs']) if data[key]['probs'] else 0
        uncertainty_avg = np.mean(data[key]['uncertainties']) if data[key]['uncertainties'] else 0
        print(f"{key.capitalize()} - Probability Average: {prob_avg:.4f}")
        print(f"{key.capitalize()} - Uncertainty Average: {uncertainty_avg:.4f}")

# 파일명에 맞는 경로를 지정하세요
filename = '/home/jykang/models/aasist/exp_result_evid/LA_AASIST_debug_ep100_bs24/eval_scores_using_best_dev_model.txt'
calculate_averages(filename)
```

```python
import numpy as np
import pandas as pd
import re

def calculate_averages_and_save_to_excel(filename, output_filename, eer_filename):
    spoof_keys = ['A07', 'A08', 'A09', 'A10', 'A11', 'A12', 'A13', 'A14', 'A15', 'A16', 'A17', 'A18', 'A19']
    data = {key: {'probs': [], 'uncertainties': [], 'EER': []} for key in ['bonafide'] + spoof_keys}

    # 파일 읽기 및 데이터 저장
    with open(filename, 'r') as file:
        for line in file:
            parts = line.split()
            if len(parts) < 5:
                continue

            category = parts[2]
            prob = float(parts[3])
            uncertainty = float(parts[4])
            spoof_t = parts[1]

            if category == 'bonafide':
                key = 'bonafide'
            elif category == 'spoof' and spoof_t.startswith('A'):
                key = spoof_t
            else:
                continue

            data[key]['probs'].append(prob)
            data[key]['uncertainties'].append(uncertainty)
    
    with open(eer_filename, 'r') as eer_file:
        for line in eer_file:
            match = re.search(r"EER\s(A\d{2})\s+=\s+([\d.]+)", line)
            if match:
                attack_type = match.group(1)
                eer_value = float(match.group(2))

                data[attack_type]['EER'] = eer_value
                

    # 평균 계산
    averages = {
        'Type': [],
        'Prob_Avg': [],
        'Uncertainty_Avg': [],
        'EER(%)': []
    }

    for key in data:
        prob_avg = np.mean(data[key]['probs']) if data[key]['probs'] else 0
        uncertainty_avg = np.mean(data[key]['uncertainties']) if data[key]['uncertainties'] else 0
        averages['Type'].append(key)
        averages['Prob_Avg'].append(prob_avg)
        averages['Uncertainty_Avg'].append(uncertainty_avg)
        averages['EER(%)'].append(data[key]['EER'])

    # # 데이터프레임 생성 및 엑셀 파일로 저장
    df = pd.DataFrame(averages)
    df.to_excel(output_filename, index=False)

    print(f"Averages have been saved to {output_filename}")

# 파일명에 맞는 경로를 지정하세요
filename = '/home/jykang/models/aasist/exp_result_evid_softplus_seed1111/LA_AASIST_debug_ep100_bs24/eval_scores_using_best_dev_model.txt'
eer_filename = '/home/jykang/models/aasist/exp_result_evid_softplus_seed1111/LA_AASIST_debug_ep100_bs24/t-DCF_EER.txt'
output_filename = 'spoofing_averages.xlsx'
calculate_averages_and_save_to_excel(filename, output_filename, eer_filename)

```
